```python
import pandas as pd
import numpy as np  
```

```python
data_df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
data_df.head(15)
```

```python
data_df.columns
```

```python
datalist = data_df.columns.tolist()
datalist
```

```python
for item in datalist:
    if data_df[item].isna().any():
        print(f"{item} has missing values")
    else:
        print(f"{item} has no missing values")
```

```python
replace_cols = [ 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                    'TechSupport','StreamingTV', 'StreamingMovies']

for col in replace_cols:
    data_df[col] = data_df[col].replace({'No internet service': 'No'})

```

```python
data_df['MultipleLines'] = data_df['MultipleLines'].replace({'No phone service': 'No'})
```

```python
data_df['Churn'].value_counts()
```

```python
data_df.columns
data_df.head()
```

```python

data_df['gender'].value_counts()
data_df['Churn']
```

```python
data_df['gender'] = data_df['gender'].replace({'Male': 1, 'Female': 0})
```

```python
data_df['Churn'] = data_df['Churn'].replace({'Yes':1, 'No':0})
data_df.head()
```

```python
encoding_list = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection','TechSupport', 'StreamingTV', 'StreamingMovies','Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'PaperlessBilling']
for col in encoding_list:
   data_df[col] = data_df[col].replace({'Yes':1, 'No':0})
```

```python
data_df.head()
data_df.shape
data_df.columns
```

```python
data_df.drop(['customerID'], axis=1, inplace=True)
```

```python
data_df.columns
```

```python
cleaned_data = data_df.copy()
cleaned_data.to_csv('../data/cleaned_telco_customer_churn.csv', index=False)
```
